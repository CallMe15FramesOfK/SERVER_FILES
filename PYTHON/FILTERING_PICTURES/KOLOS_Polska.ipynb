```python
hsv = cv2.cvtColor(mapa, cv2.COLOR_BGR2HSV)

cv2_imshow(hsv)

maskO = cv2.inRange(hsv, (20, 100, 100), (30, 255,255))
maskP = cv2.inRange(hsv, (160, 0, 0), (180, 255, 255))

maskOP = cv2.add(maskP,maskO)

mapaF = cv2.bitwise_and(mapa,mapa, mask=maskOP)
cv2_imshow(mapaF)
```
