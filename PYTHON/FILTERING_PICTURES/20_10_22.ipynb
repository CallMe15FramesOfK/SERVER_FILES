```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import skimage,imutils
```

```python
img = cv2.imread('/mydrive/Zdjęcia(Moje)/me.png')
print(img.shape)
print(type(img))
cv2_imshow(img)
img2 = cv2.imread('/mydrive/Zdjęcia(Moje)/go.png')

```

```python
scale = 1.5

rows, cols = img.shape[:2]
row2 = int(rows* scale)
col2 = int(cols * scale)


img_s = cv2.resize(img, (col2,row2))
cv2_imshow(img_s)
```

```python
#obracanie


img_r = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

cv2_imshow(img_r)
```

```python
import imutils

img_ro = imutils.rotate_bound(img,45)
cv2_imshow(img_ro)
img_ro2 = imutils.rotate(img,40)
cv2_imshow(img_ro2)
```

```python
img_c = img + 40
cv2_imshow(img_c)
```

```python
import numpy as np

alpha = 1.0  #zakres od 0.0 do 2.0
beta = -50   #zakres od -255 do 255

def jasnosc(a,b,img):
  return np.clip(alpha * img + beta, 0,255)


img_c = jasnosc(alpha,beta,img)
cv2_imshow(img_c)  


```
